```python
# SIMPLE LINEAR REGRESSION

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


dataset = pd.read_csv("placement.csv")
dataset = dataset.dropna()
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["cgpa"]]  # we will convert cgpa 1 dimension into 2 dimension by adding square brackets
y = dataset[["package"]]
```

```python
# TRAINING DATA

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 50)

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# ACCURACY OF MODEL

lr.score(x_test,y_test)*100

# IN ORDER TO CHANGE THE ACCURACY OF THE MODEL CHANGE THE VALUE OF RANDOM_STATE
```

```python
# PREDICT PACKAGE

lr.predict([[6.89]])
```

```python
# to find c i.e intercept

lr.intercept_
```

```python
# y = m*x +c 

lr.coef_    # m i.e slope
```

```python
y_prd = lr.predict(x)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.plot(dataset["cgpa"],y_prd,c="red")
plt.legend(["org data","prediction line"])
plt.title("PACKAGE PREDICTION GRAPH")
plt.show()
```
