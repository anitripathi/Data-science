```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("Placement (1).csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
# Inorder to check if dataset is normally distributed

sns.kdeplot(data=dataset["CGPA"])
plt.show()
```

```python
sns.scatterplot(x="CGPA",y="IQ",data=dataset,hue="Placement")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Placement"]

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
# to apply naive bayes algorithm

from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
from sklearn.decomposition import PCA

# Reduce features to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x)

# Ensure y is numeric
y = y.to_numpy().astype(int)

# Train the model
gnb.fit(X_pca, y)

# Plot the decision regions
plot_decision_regions(X_pca, y, clf=gnb)
plt.title("Decision Boundary (Naïve Bayes)")
plt.show()
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_train,y_train)*100,mnb.score(x_test,y_test)*100
```

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_train,y_train)*100,bnb.score(x_test,y_test)*100
```
