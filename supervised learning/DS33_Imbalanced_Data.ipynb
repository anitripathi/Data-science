```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

dataset = pd.read_csv("social.csv")
dataset.head(5)
```

```python
dataset["Purchased"].value_counts()
```

```python
# Encode Categorical Feature (Before Splitting)
encoder = LabelEncoder()
dataset['Gender'] = encoder.fit_transform(dataset['Gender'])  # Male → 1, Female → 0
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]

```

```python
# Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
# Train Logistic Regression Model

lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
# Predictions
y_pred = lr.predict(x_test)
print("Predictions:", y_pred)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[15624510,1,29.0,19000.0]])
```

 this model is biased

```python
# now we will use inblean module to balance the biased dataset

from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
```

```python
ru_x,ru_y = ru.fit_resample(x,y)
```

```python
ru_y
```

```python
ru_y.value_counts()   # we can observe that our dataset is balanced
```

```python
x_train, x_test, y_train, y_test = train_test_split(ru_x, ru_y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[15798659,1,80.00,69000.00]])
```
