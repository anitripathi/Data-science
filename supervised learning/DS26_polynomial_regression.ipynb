```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)  # Independent variable
y = 2 * X**3 - 5 * X**2 + 3 * X + np.random.randn(100, 1) * 5  # Non-linear relation with noise
```

```python
poly = PolynomialFeatures(degree=3)  # Choose the degree of polynomial
X_poly = poly.fit_transform(X)  # Transform X to include polynomial terms
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Polynomial Regression')
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression Example")
plt.show()
```

```python
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```
