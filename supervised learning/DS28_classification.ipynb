```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("Boston.csv")
data_sel = dataset[['chas','crim']]

data_sel.head(3)
```

```python
sns.scatterplot(x="crim",y="chas",data=data_sel)
plt.show()
```

```python
x=data_sel[["crim"]]
y = data_sel[["chas"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,t_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
lr.fit(data_sel[["crim"]], data_sel["chas"])
```

```python
data_sel["predicted_chas"] = lr.predict(data_sel[["crim"]])
```

```python
lr.score(x_train,y_train)*100
```

```python
lr.predict([[0.09]])
```

```python
sns.scatterplot(x="crim", y="chas", data=data_sel, label="Actual Data")
sns.lineplot(x="crim", y="predicted_chas", data=data_sel, color="red", label="Regression Line")
plt.show()
```
