```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

# Generate synthetic dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(1000, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])  # Sine function with noise

# Define models
bagging = BaggingRegressor(estimator=DecisionTreeRegressor(), n_estimators=50, random_state=42)
random_forest = RandomForestRegressor(n_estimators=50, random_state=42)

# Fit models
bagging.fit(X, y)
random_forest.fit(X, y)

# Predictions
X_test = np.linspace(0, 5, 1000).reshape(-1, 1)
y_bagging_pred = bagging.predict(X_test)
y_rf_pred = random_forest.predict(X_test)

# Plot results
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color="gray", label="Training Data", alpha=0.3)

plt.plot(X_test, y_rf_pred, color="red", linestyle="dashed", label="Random Forest Predictions")

plt.title("Comparison of Bagging and Random Forest")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
