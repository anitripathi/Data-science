```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("irisraw.csv")
```

```python
dataset.head(5)
```

```python
dataset = dataset.drop(columns=['species'])
```

```python
dataset.head(5)
```

```python
# to check if dataset is linearly seperable

# sns.pairplot(data=dataset)
# plt.show()
```

```python
# dendrogram (pip install scipy)

import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,10))
sc.dendrogram(sc.linkage(dataset,method='single',metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
```

```python
dataset["Predict"]=ac.fit_predict(dataset)
```

```python
dataset
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.show()
```
