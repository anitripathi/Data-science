```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generate sample dataset (moons shape)
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred = dbscan.fit_predict(X)

# Plot the clusters
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s=50, color='red', label='Cluster 1')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s=50, color='blue', label='Cluster 2')
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], s=50, color='black', label='Outliers') # Noise points
plt.legend()
plt.title("DBSCAN Clustering")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons    # take inbult non linear dataset
```

```python
x,y = make_moons(n_samples=250,noise=0.05)
```

```python
df = {"data1" : x[:,0],"data2":x[:,1]}
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2,min_samples=5)
dataset["Predict"] = db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1",y="data2",data=dataset,hue="Predict")
plt.show()
```
