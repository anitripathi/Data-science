```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("loan.csv")
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
# observe outliers through distribution plot

sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
# DIRECT METHOD :-

min_range = dataset["CoapplicantIncome"].mean() - (3*dataset["CoapplicantIncome"].std())
max_range = dataset["CoapplicantIncome"].mean() + (3*dataset["CoapplicantIncome"].std())
```

```python
min_range,max_range
```

```python
new_data = dataset[dataset["CoapplicantIncome"]<=max_range]

```

```python
new_data.shape
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x = "CoapplicantIncome",data=new_data)  
plt.show()
```

```python
# Z-SCORE METHOD

z_score = (dataset["CoapplicantIncome"] - dataset["CoapplicantIncome"].mean())/dataset["CoapplicantIncome"].std() 
```

```python
dataset["z_score"] = z_score   # add 1 column in datset of name z_score
```

```python
dataset[dataset["z_score"]<3]   # you can observe no. of rows decreased as outlier are removed using z-score
```
