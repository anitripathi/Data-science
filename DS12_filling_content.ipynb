```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("Toyota.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
# Fills 10 where data is missing

# dataset.fillna(10)
```

```python
# BACKWARD FILLING 
dataset.bfill()


# FORWARD FILLING

# dataset.ffill()
```

```python
dataset["Age"].mode()[0]     # Fill the most occured or ode data in missing place
dataset["Age"].fillna(dataset["Age"].mode()[0],inplace=True)   # Replace missing in dataset with mode data in age column

```

```python
dataset.info()
```

```python
# to fill mode value in OBJECT type data column in dataset

for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# to fill mode value in INTEGER type data column in dataset

for i in dataset.select_dtypes(include="float64").columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```
