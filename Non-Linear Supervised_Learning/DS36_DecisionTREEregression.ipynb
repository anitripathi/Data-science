```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([600, 800, 1000, 1500, 1800]).reshape(-1, 1)
y = np.array([30, 45, 55, 75, 90])  # Price in Lakhs

# Create and train the model
model = DecisionTreeRegressor(max_depth=2)
model.fit(X, y)

# Predict on new data
X_test = np.linspace(500, 2000, 100).reshape(-1, 1)
y_pred = model.predict(X_test)

# Plot results
plt.scatter(X, y, color='red', label="Actual Prices")
plt.plot(X_test, y_pred, color='blue', label="Regression Tree Prediction")
plt.xlabel("Square Feet")
plt.ylabel("Price in Lakhs")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Generate 1000 random data points for regression
np.random.seed(42)  # For reproducibility
X = np.sort(10 * np.random.rand(1000, 1), axis=0)  # Feature values between 0 and 10
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])  # Target with noise

# Define different max_depth values for comparison
depths = [2, 4, 6, None]  # None means no limit (overfitting)

plt.figure(figsize=(12, 8))

for i, depth in enumerate(depths, 1):
    model = DecisionTreeRegressor(max_depth=depth)
    model.fit(X, y)

    # Generate predictions
    X_test = np.linspace(0, 10, 1000).reshape(-1, 1)
    y_pred = model.predict(X_test)

    # Plot results
    plt.subplot(2, 2, i)
    plt.scatter(X, y, color='red', s=5, alpha=0.5, label="Actual Data")
    plt.plot(X_test, y_pred, color='blue', linewidth=2, label=f"Tree Depth = {depth}")
    plt.xlabel("Feature (X)")
    plt.ylabel("Target (y)")
    plt.title(f"Decision Tree Regression (Depth = {depth})")
    plt.legend()

plt.tight_layout()
plt.show()
```
