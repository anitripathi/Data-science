```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Salary.csv")
```

```python
dataset.head(3)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# dt = DecisionTreeRegressor()
dt = DecisionTreeRegressor(criterion='friedman_mse', max_depth=6, splitter='random')   # hmne best estimator check kr k yha changes lia hai jisse accuracy increase ho shake  
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100
```

```python
dt.score(x_test,y_test)*100
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python

df = {
    "criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter": ["best", "random"],
    "max_depth": [i for i in range(2, 20)]  # Corrected list comprehension syntax
}

# Initialize Decision Tree Regressor
regressor = DecisionTreeRegressor()

# Use GridSearchCV to find the best parameters
grid_search = GridSearchCV(estimator=regressor, param_grid=df, cv=5, scoring='r2')
```

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid=df, cv=5, scoring='r2', n_jobs=-1)
```

```python
gd.fit(x_train,y_train)
```

```python
# Print the best parameters
print("Best Parameters:", gd.best_params_)

# Print the best score
print("Best Score:", gd.best_score_*100)
```
