```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("loan.csv")

# Display the first few rows
df.head(3)
```

```python
df[['LoanAmount', 'ApplicantIncome']].skew()
```

```python
#  LOG TRANSFORMATION  (Used to normalize right-skewed data like LoanAmount)  


# Create a figure with 2 subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # (rows, columns), figure size

sns.histplot(df['LoanAmount'], kde=True, ax=axes[0])
axes[0].set_title("Before Log Transformation")

df['Log_LoanAmount'] = np.log1p(df['LoanAmount'])  # log1p avoids log(0)
sns.histplot(df['Log_LoanAmount'], kde=True, ax=axes[1])
axes[1].set_title("After Log Transformation")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# SQUARE ROOT TRANSFORMATION  (Used when data is moderately skewed)

df['Sqrt_ApplicantIncome'] = np.sqrt(df['ApplicantIncome'])
sns.histplot(df['Sqrt_ApplicantIncome'], kde=True)
plt.show()
```

```python
# LAMDA FUNCTION TRANSFORMATION


# Before transformation plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df['ApplicantIncome'], kde=True, ax=axes[0])
axes[0].set_title("Before Square Root Transformation")

# Apply Square Root transformation using lambda function
df['Sqrt_ApplicantIncome'] = df['ApplicantIncome'].apply(lambda x: np.sqrt(x))

# After transformation plot
sns.histplot(df['Sqrt_ApplicantIncome'], kde=True, ax=axes[1])
axes[1].set_title("After Square Root Transformation")

plt.tight_layout()
plt.show()
```
