```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a Skewed Population (Exponential Distribution)
population = np.random.exponential(scale=5, size=10000)  # Mean ~ 5

# Step 2: Draw Random Samples & Compute Sample Means
sample_sizes = [5, 10, 30, 50]  # Different sample sizes
sample_means = {}

for n in sample_sizes:
    means = [np.mean(np.random.choice(population, size=n, replace=True)) for _ in range(1000)]
    sample_means[n] = means  # Store sample means for plotting

# Step 3: Plot the Results
plt.figure(figsize=(12, 8))
for n, means in sample_means.items():
    sns.kdeplot(means, label=f"Sample Size = {n}")

plt.axvline(x=np.mean(population), color='r', linestyle='dashed', label="Population Mean")
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.show()
```
