```python
# HANDALING MISSING VALUE

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("Toyota.csv")
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull())   #  WHITE BARS SHOWS MISSING VALUE
plt.show()
```

```python
# column deletion

# METHOD 1
dataset.drop(columns=["MetColor"],inplace=True)   # METCOLOR IS DELETED AND NEW DATASET OF 10 COLUMN IS CREATED
dataset.head(5)
```

```python
dataset.shape
```

```python
# ROW DELETION IS ANY MISSING CONTENT IS PRESENT

dataset.dropna(inplace = True)
```

```python
dataset.shape
```

```python
sns.heatmap(dataset.isnull())   #  ALL MISSING VALUE IS DELETED 
plt.show()
```
