```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mu, sigma = 0, 1  # Standard normal distribution

# Generate x values
x = np.linspace(-4, 4, 1000)

# Compute PDF
pdf = norm.pdf(x, mu, sigma)  #normal distribution function

# Plot PDF
plt.plot(x, pdf, label='Normal Distribution', color='blue')
plt.title("Normal Distribution Curve")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
import numpy as np

# Sample data (Math and Science scores)
math_scores = [50, 60, 70, 80, 90]
science_scores = [60, 65, 70, 85, 95]

# Compute correlation coefficient
correlation = np.corrcoef(math_scores, science_scores)[0, 1]

print("Correlation between Math and Science scores:", correlation)
```

```python
import numpy as np

# Sample data (Math and Science scores)
math_scores = [50, 60, 70, 80, 90]
science_scores = [60, 65, 70, 85, 95]

# Compute covariance matrix
cov_matrix = np.cov(math_scores, science_scores, bias=True)

# Extract covariance value
covariance_value = cov_matrix[0, 1]

print("Covariance between Math and Science scores:", covariance_value)
```

```python
#COR-RELATION

data_cor=df.select_dtypes(["int64","float64"]).corr()   
data_cor
```

```python
#COVARIANCE

data_cov=df.select_dtypes(["int64","float64"]).cov() 
data_cov
```

```python
sns.heatmap(data_cor)
plt.title("CORRELATION HEATGRAPH")
plt.show()
```

```python
sns.heatmap(data_cov,annot=True)

plt.title("COVARIANCE HEATGRAPH")
plt.show()
```
