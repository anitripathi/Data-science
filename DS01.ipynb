```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Toyota.csv")
# print(df.head(5))
# print(df.info())          
mean_price = np.mean(df["Age"])
print(mean_price)
plt.hist(df["Age"],bins=10,color='red',edgecolor='black')
plt.plot([mean_price for i in range(0,200)],[i for i in range(0,200)],c="yellow")
plt.title("MEAN")
# median_value= df["Age"].dropna().median()
# print(median_value)
plt.show()
```

```python
# MEDIAN
median_value= df["Age"].dropna().median()
print(median_value)
plt.hist(df["Age"],bins=10,color='blue',edgecolor='black')
plt.plot([median_value for i in range(0,500)],[i for i in range(0,500)],c="yellow")
plt.title("MEDIAN")
plt.show()
```

```python
# MODE
# print(df.head())
md=df["Age"].mode()
plt.hist(df["Age"],bins=10,color='aqua',edgecolor='black')
plt.plot([md for i in range(0,300)],[i for i in range(0,300)],c="red")
plt.title("MODE")
print(df["Age"].value_counts())       #counts the no. of freguency
plt.show()
```
